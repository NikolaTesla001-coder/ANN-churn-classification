```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#load the model
model = load_model('model.h5')

#load the scaler and encoder
with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
with open('OHE.pkl', 'rb') as f:
    onehot_encoder = pickle.load(f)
with open('label.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
#one hot encode the Geography column
ip_geo=onehot_encoder.transform([[input_data['Geography']]]).toarray()
ip_geo_df=pd.DataFrame(ip_geo,columns=onehot_encoder.get_feature_names_out(['Geography']))
```

```python
##dataframe for input data
df_input=pd.DataFrame([input_data])
df_input
```

```python
#label encode for the gender column
df_input['Gender']=label_encoder.transform(df_input['Gender'])
```

```python
print(label_encoder.classes_)
```

```python
print(type(onehot_encoder))
print(type(label_encoder))
print(type(scaler))
```

```python
ip_geo_df
```

```python
#concating one hot encoder cols(df_geo) and df
df=pd.concat([df_input.drop('Geography',axis=1),ip_geo_df],axis=1)
```

```python
df
```

```python
#scale the data
scaled_data=scaler.transform(df)
scaled_data
```

```python
##predicting churn
pred=model.predict(scaled_data)
```

```python
pred
```

```python
if pred>0.5:
    print('the person is not likey to leave')
else:
    print('the person is likey leave')
```
